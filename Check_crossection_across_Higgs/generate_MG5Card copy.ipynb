```python
import os
import subprocess
mh_list=[]
xsec_list=[]
for folder in os.listdir():
    path=os.path.join(os.getcwd(), folder, 'eeToZH_HLFV_FourLeptons','Events','formal01','formal01_tag_1_banner.txt')
    # os.system("grep '# mh' *.txt >> ../../../mh.txt")
    # os.system("grep '#  Integrated weight' *.txt >> ../../../xsec.txt")
    mh=subprocess.run(['grep', '# mh', path], text=True, capture_output=True).stdout
    xsec=subprocess.run(['grep', '#  Integrated weight', path], text=True, capture_output=True).stdout
    mh_list.append(float(mh.split(" ")[-3]))
    try:
        xsec_list.append(float(xsec.split(" ")[-1][:-1]))
    except:
        xsec_list.append(0)
    # os.chdir("..")
```

```python
sorted_pairs = sorted(zip(mh_list, xsec_list))
sorted_mh, sorted_xsec = zip(*sorted_pairs)
```

```python
sorted_xsec
```

```python
import matplotlib.pyplot as plt
plt.plot(sorted_mh, sorted_xsec)
plt.xlabel('Higgs Mass (GeV)')
plt.ylabel('Cross Section (pb)')
plt.title('Higgs Mass vs eeToHZ_Zll Cross Section')
# plt.yscale('log')
plt.grid()
plt.show()
```

```python
plt.plot(sorted_mh, sorted_xsec)
plt.xlabel('Higgs Mass (GeV)')
plt.ylabel('Cross Section (pb)')
plt.title('Higgs Mass vs eeToHZ_Zll Cross Section')
plt.yscale('log')
plt.grid()
plt.show()
```

```python
# now plot as previously done but mark each data point and add the value next to it
# text value situated at left side of the point
# value written as scientific notation with 2 decimal places
plt.plot(sorted_mh, sorted_xsec, marker='o', linestyle='-')
for i, txt in enumerate(sorted_xsec):
    plt.annotate(f"{txt:.2e}", (sorted_mh[i], sorted_xsec[i]), textcoords="offset points", xytext=(0,-10), ha='center', fontsize=8)
plt.xlabel('Higgs Mass (GeV)')
plt.ylabel('Cross Section (pb)')
plt.title('Higgs Mass vs eeToHZ_Zll Cross Section')
plt.yscale('log')
plt.grid()
plt.show()
```
